```
# Import necessary packages
import numpy as np
import pandas as pd
import time
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```
# Scrape data from the website
score = pd.DataFrame(columns = ['Name', 'Birth_of_date', 'Content'])
for i in range(2068001, 2074719): #2074719
    link_href = 'http://diemthi.hcm.edu.vn/'
    driver.get(link_href)
    
    # Fill in SBD:
    sbd = driver.find_element_by_xpath('//*[@id="SoBaoDanh"]')
    sbd.send_keys('0' + str(i))
    button = driver.find_element_by_xpath('//*[@id="loginForm"]/form/div/input[2]')
    button.click()

    # Get data
    try:
        name = driver.find_element_by_xpath('/html/body/div/table/tbody/tr[2]/td[1]').text
    except:
        score.loc[len(score)] = [np.nan, np.nan, np.nan]
        driver.get(link_href)
        continue

    birth_of_date = driver.find_element_by_xpath('/html/body/div/table/tbody/tr[2]/td[2]').text
    content = driver.find_element_by_xpath('/html/body/div/table/tbody/tr[2]/td[3]').text

    score.loc[len(score)] = [name, birth_of_date, content]
    print(i)
    driver.get(link_href)
```
